## Q4

```python
import IPython.display as ipd
```

```python
sr = 44100
t = 1
T = int(sr / 100)

sig = np.linspace(0, t, t * sr)

# random amplitudes
a1 = np.random.random(1)
a2 = np.random.random(1)
a3 = np.random.random(1)

# sinewaves
s1 = np.sin(2 * np.pi * 100 * sig) * a1
s2 = np.sin(2 * np.pi * 200 * sig) * a2
s3 = np.sin(2 * np.pi * 300 * sig) * a3

# mixture
mix = s1 + s2 + s3
```

```python
ipd.display(ipd.Audio(s1, rate=sr))

plt.plot(s1[:T])
plt.title('100Hz')
plt.show()
```

```python
ipd.display(ipd.Audio(s2, rate=sr))
plt.plot(s2[:T])
plt.title('200Hz')
plt.show()
```

```python
ipd.display(ipd.Audio(s3, rate=sr))
plt.plot(s3[:T])
plt.title('300Hz')
plt.show()
```

```python
ipd.display(ipd.Audio(mix, rate=sr))
plt.plot(mix[:T])
plt.title('Mixture')
plt.show()
```

```python
# Amplitude Estimation
freqs = np.array([s1 / a1, s2 / a2, s3 / a3])

print('Amplitude estimation with base signals')
print('Generated:', a1, a2, a3)
print('Estimated:', np.dot(freqs, mix) / sr * 2)
```

```python
# random amplitudes
a1 = np.random.random(1)
a2 = np.random.random(1)
a3 = np.random.random(1)

# sinewaves
s1 = np.sin(2 * np.pi * 100 * sig) * a1
s2 = np.sin(2 * np.pi * 200 * sig) * a2
s3 = np.sin(2 * np.pi * 300 * sig) * a3

# random imaginary component
i1 = np.random.random(s1.size)
i2 = np.random.random(s1.size)
i3 = np.random.random(s1.size)

# imaginary --> complex
p1 = np.array([])
for n in range(s1.size):
  p1 = np.append(p1, complex(0, i1[n]))

p2 = np.array([])
for n in range(s1.size):
  p2 = np.append(p2, complex(0, i2[n]))

p3 = np.array([])
for n in range(s1.size):
  p3 = np.append(p3, complex(0, i3[n]))

r1 = np.array([x+0j for x in s1])
r2 = np.array([x+0j for x in s2])
r3 = np.array([x+0j for x in s3])

s1 = r1 + p1
s2 = r2 + p2
s3 = r3 + p3

# mixture
mix = s1 + s2 + s3
```

```python
ipd.display(ipd.Audio(s1, rate=sr))

plt.plot(s1[:T])
plt.title('100Hz')
plt.show()
```

```python
ipd.display(ipd.Audio(s2, rate=sr))
plt.plot(s2[:T])
plt.title('200Hz')
plt.show()
```

```python
ipd.display(ipd.Audio(s3, rate=sr))
plt.plot(s3[:T])
plt.title('300Hz')
plt.show()
```

```python
ipd.display(ipd.Audio(mix, rate=sr))
plt.plot(mix[:T])
plt.title('Mixture')
plt.show()
```

```python
# Amplitude Estimation - Sine
x1 = np.sin(2 * np.pi * 100 * sig)
x2 = np.sin(2 * np.pi * 200 * sig)
x3 = np.sin(2 * np.pi * 300 * sig)

xs = np.array([x1, x2, x3])

print('Amplitude estimation with sine')
print('Generated:', a1, a2, a3)
print('Estimated:', (np.dot(xs, mix) / sr * 2).real)

# Phase Estimation - Cosine
x1 = np.cos(2 * np.pi * 100 * sig)
x2 = np.cos(2 * np.pi * 200 * sig)
x3 = np.cos(2 * np.pi * 300 * sig)

xs = np.array([x1, x2, x3])

print('\nPhase estimation with cosine')
print('Generated:', i1, i2, i3)
print('Estimated:', (np.dot(xs, mix) / sr * 2).imag)
```

## Q7

```python
def n_temperament_scale(n, f, sr):

  x = np.linspace(0, 1, sr)

  # ADSR params
  attack_length = 0.1
  decay_length = 0.1
  sustain_length = 0.75
  release_length = 0.05
  sustain_level = 0.8

  attack = np.linspace(0, 1, int(sr * attack_length))
  decay = np.linspace(1, sustain_level, int(sr * decay_length))
  sustain = np.full(int(sr * sustain_length), sustain_level)
  release = np.linspace(sustain_level, 0, int(sr * release_length))

  envelope = attack
  envelope = np.append(envelope, decay)
  envelope = np.append(envelope, sustain)
  envelope = np.append(envelope, release)

  sig = np.array([])

  # calculating pitches
  for i in range(1, n+1):
    pitch = f * (2 ** (1/n)) ** (i-1)
    to_add = np.array([pitch for j in range(sr)])
    to_add = np.sin(2 * np.pi * pitch * x)
    to_add *= envelope
    sig = np.append(sig, to_add)

  for i in range(n-1, 0, -1):
    pitch = f * (2 ** (1/n)) ** (i-1)
    to_add = np.array([pitch for j in range(sr)])
    to_add = np.sin(2 * np.pi * pitch * x)
    to_add *= envelope
    sig = np.append(sig, to_add)

  # add bass drone
  bass_sustain_length = sig.size - ((attack_length + decay_length + release_length) * sr)
  bass_sustain = np.full(int(sr * bass_sustain_length), sustain_level)

  bass_envelope = attack
  bass_envelope = np.append(bass_envelope, decay)
  bass_envelope = np.append(bass_envelope, bass_sustain)
  bass_envelope = np.append(bass_envelope, release)

  y = np.linspace(0, 2 * n - 1, (2 * n - 1) * sr)
  sig += np.sin(2 * np.pi * f * y)
  sig *= bass_envelope

  ipd.display(ipd.Audio(sig, rate=sr))
```

```python
sr = 44100
f = 110     # bass frequency

for n in range(3, 13):
  n_temperament_scale(n, f, sr)
```

## Q8

```python
import math
```

```python
def clock(min=0, sec=0):
  min_x = math.sin(2 * math.pi * min/60)
  min_y = math.cos(2 * math.pi * min/60)

  sec_x = math.sin(2 * math.pi * sec/60)
  sec_y = math.cos(2 * math.pi * sec/60)

  return min_x, sec_x, min_y, sec_y
```

```python
min_x, sec_x, min_y, sec_y = clock(10, 30)

figure, axes = plt.subplots(figsize=[5, 5])
circle = plt.Circle((0, 0), 1, fill=False)
axes.add_artist(circle)

plt.plot([0, min_x], [0, min_y], label='minutes')
plt.plot([0, sec_x], [0, sec_y], label='seconds')

plt.xlim([-1.1, 1.1])
plt.ylim([-1.1, 1.1])
axes.xaxis.set_visible(False)
axes.yaxis.set_visible(False)
plt.legend()
plt.show()
```
